```python
import tweepy

consumer_key = " "
consumer_secret = " "
access_key = " " 
access_secret = " "

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth)
```

Finding tweets by using @mentions used in the tweets since particular date.

```python
mention = input("Enter the id you want to check for(format: @xyz):" )
date = input("Enter the date from where you want to find(format: yyyy-mm-dd): ")
mention_tweets = tweepy.Cursor(api.search, q=mention,since=date, tweet_mode='extended').items(100)
for tweet in mention_tweets:
  print(tweet._json['full_text'])
```

Finding tweets using #hashtag used in particular tweets since particular date.

```python
hashtag = input("Enter the hashtag you want to check for(format: #xyz): ")
date = input("Enter the date from where you want to find(format: yyyy-mm-dd): ")
hashtag_tweets = tweepy.Cursor(api.search, q=hashtag, since=date, tweet_mode='extended').items(100)

for tweet in hashtag_tweets:
  print(tweet._json['full_text'])
```

Find tweets of particular user.

```python
username = input("Enter the username to extract it's tweets(format: xyz): ")
new_tweets = tweepy.Cursor(api.user_timeline, screen_name=username, tweet_mode='extended').items(100)

tweets = []
for tweet in new_tweets:
  tweets.append(tweet._json)

tweets[:len(tweets)]
```

Extract all tweets of a particular user from their very first tweet.

```python
all_content = tweepy.Cursor(api.user_timeline, screen_name=username, tweet_mode='extended').pages()
all_tweets = []
for page in all_content:
    for tweet in page:
        all_tweets.append(tweet._json)

print("Number of tweets are:", len(all_tweets))
```

```python
import re
def handle_tweet_text(tweet_text):

    for url in re.findall(r'(https?://\S+)', tweet_text):
        tweet_text = tweet_text.replace(url, f'<a href="{url}">{url}</a>')

    if len(tweet_text) >= 356:
        tweet_text = f'{tweet_text[:350]} . . .'

    return tweet_text
```

```python
all_tweets_refined = []
for tweet in all_tweets:
    temp = {}
    temp['id'] = tweet['id']
    #fetch created time
    temp['created_at'] = tweet['created_at']
    #fetch text
    temp['text'] = handle_tweet_text(tweet['full_text'])
    #fetch hashtags for the tweet
    try:
        hashtags_list = tweet['entities']['hashtags']
        hashtags = [hashtag['text'] for hashtag in hashtags_list]
        if len(hashtags_list)>1:
            temp['hashtags'] = ",".join(hashtags)
        else:
            temp['hashtags'] = hashtags_list[0]['text']
    except:
        temp['hashtags'] = ''
    #fetch user mentions
    mentions_list = [mention['screen_name'] for mention in tweet['entities']['user_mentions']]
    temp['mentions'] = ",".join(mentions_list)
    #fetch url for the tweet
    try:
        temp['url'] = tweet['entities']['urls'][0]['url']
    except:
        temp['url'] = ''
    #Fetch location (if available) for tweet
    try:
        temp['location'] = tweet['user']['location']
    except:
        temp['location'] = None
    #fetch number of times tweet has been retweeted
    temp['retweet_count'] = tweet['retweet_count']
    all_tweets_refined.append(temp)
```

Tabulating all the tweets made by an account till date.

```python
import pandas as pd
tweets_df = pd.DataFrame(all_tweets_refined)
tweets_df['created_at'] = pd.to_datetime(tweets_df['created_at'])
tweets_df['created_at'] = tweets_df['created_at'].map(lambda t: t.tz_convert(None))
tweets_df.head(len(all_tweets))
```
